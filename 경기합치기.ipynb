```python
import numpy as np
import pandas as pd
import os
os.getcwd()
```

```python
Seocho_Tem=pd.read_csv('용산_시간별.csv', engine='python', index_col=0)

GG=pd.read_csv('Gyeonggi.csv',engine='python', index_col=0)

GG.head()

Seocho_Tem.tail()

Seocho_Tem = Seocho_Tem[['풍속','기온','강수','습도','Time']]

GG['Time']=pd.to_datetime(GG['Time'])

Seocho_Tem['Time']=pd.to_datetime(Seocho_Tem['Time'])

final=pd.merge(Seocho_Tem,GG, on='Time')

final['Label'] = final['기온'] - final['Temperature']

final['label']=final['Label']>3

final.label=final.label.astype('int')

print(len(final[final.label==1]))
print(len(final[final.label==0]))

final['Time'] = final.Time.astype('str')

final['Season']=final.Time.str.split('-',expand=True).iloc[:,1]

#F=final[final.label==1]
s_1=final[final['Season']=='01'].index
s_2=final[final['Season']=='02'].index
s_3=final[final['Season']=='03'].index
s_4=final[final['Season']=='04'].index
s_5=final[final['Season']=='05'].index
s_6=final[final['Season']=='06'].index
s_7=final[final['Season']=='07'].index
s_8=final[final['Season']=='08'].index
s_9=final[final['Season']=='09'].index
s_10=final[final['Season']=='10'].index
s_11=final[final['Season']=='11'].index
s_12=final[final['Season']=='12'].index


for i in range(len(s_1)):
    final.at[s_1[i],'Season'] = 'Winter'
for i in range(len(s_2)):
    final.at[s_2[i],'Season'] = 'Winter'
for i in range(len(s_3)):
    final.at[s_3[i],'Season'] = 'Spring'
for i in range(len(s_4)):
    final.at[s_4[i],'Season'] = 'Spring'
for i in range(len(s_5)):
    final.at[s_5[i],'Season'] = 'Spring'
for i in range(len(s_6)):
    final.at[s_6[i],'Season'] = 'Summer'
for i in range(len(s_7)):
    final.at[s_7[i],'Season'] = 'Summer'
for i in range(len(s_8)):
    final.at[s_8[i],'Season'] = 'Summer'
for i in range(len(s_9)):
    final.at[s_9[i],'Season'] = 'Autumn'
for i in range(len(s_10)):
    final.at[s_10[i],'Season'] = 'Autumn'
for i in range(len(s_11)):
    final.at[s_11[i],'Season'] = 'Autumn'
for i in range(len(s_12)):
    final.at[s_12[i],'Season'] = 'Winter'

print(len(final))
print(len(final.dropna()))

final.to_csv('용산_경기.csv',encoding='cp949')

final.head()
```
